```python
temp1 = soup.find_all("tr")[1]

def _trade_info(cellSoup, tradeType):

    children = [i for i in cellSoup.children][1::2]
    
    info = dict([i.strip().split(" ") for i in children[1].text.strip().split("|")])
    info[u"Rating"] = float(info[u"Rating"][:-1])/100
    info[u"Trades"] = int(info[u"Trades"])
    info[u"Trusted"] = int(info[u"Trusted"])

    info_text_list = [i.text.strip() for i in children]
    
    info[u"User"] = info_text_list[0]
    info[u"UserID"] = cellSoup.find("a").attrs["href"].split("/")[2]
    info[u"Method"] = info_text_list[2]
    info[u"AmountLeast"] = float(info_text_list[3].split("-")[0])
    info[u"AmountMost"] = float(info_text_list[3].split(" ")[0].split("-")[1])
    info[u"Price"] = float(info_text_list[4].split(" ")[0])
    info[u"Type"] = tradeType

    return info

def trade_info_ask(cellSoup):
    return _trade_info(cellSoup, "ask")

def trade_info_bid(cellSoup):
    return _trade_info(cellSoup, "bid")
```
